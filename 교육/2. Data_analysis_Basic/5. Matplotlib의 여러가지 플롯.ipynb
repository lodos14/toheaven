```python
# 바 차트
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 바 차트 : bar, barh

y = [2, 3, 1]
x = np.arange(len(y))

xlabel = ['A', 'B', 'C']

plt.title('Bar Chart')

plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(y)

# 축제목
plt.xlabel('Word')
plt.ylabel('Fre')

plt.show()
```

```python
people = ['aaa', 'bbb', 'ccc', 'ddd']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel("x-label")
plt.show()
```

```python
# pie : 파이 차트

labels = ['aaa', 'bbb', 'ccc', 'ddd']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue','lightcoral']
explode = (0, 0.1, 0, 0)

plt.title('Pie Chart')
plt.pie(sizes, labels=labels, colors = colors,
       explode=explode, startangle = 90, shadow=True,
       autopct='%1.1f%%')
# 파이차트를 그릴 때 원의 형태가 유지될 수 있도록 설정
plt.axis('equal')
plt.show()
```

```python
# 히스트그램 : 빈도수, hist
# 데이터를 집계할 구간 - bins
x = np.random.randn(1000)

plt.title('Histogram')
plt.hist(x, bins = 10)
plt.show()
```

```python
# 산점도 or 산포도 or scatter, 2차원 데이터
# 두 집합의 상관관계를 나타낼 때 
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

plt.title('Scatter Plot')
plt.scatter(x, y)
plt.show()
```
